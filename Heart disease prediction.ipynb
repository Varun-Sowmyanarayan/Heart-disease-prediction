```python
df = pd.read_csv("C:/Users/91875/Downloads/Heart.csv")
```

```python
df
```

```python
df=df.drop(columns='Unnamed: 0')
df
```

```python
#convert the categorical values into numerical values
df['ChestPain']=df['ChestPain'].astype('category')
df['ChestPain']=df['ChestPain'].cat.codes
df
```

```python
df['Thal']=df['Thal'].astype('category')
df['Thal']=df['Thal'].cat.codes

df['AHD']=df['AHD'].astype('category')
df['AHD']=df['AHD'].cat.codes
df
```

```python
#check for null values
df.isnull().sum()
```

```python
#drop the null value columns
df=df.dropna()
df
```

```python
#seperating the columns for the x and y axis
X=df.drop(columns="AHD")
X
```

```python
#for y axis
y=df['AHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split the train and test data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression(random_state=0).fit(X_train_scaled,y_train)
```

```python
log_reg.predict(X_train_scaled)
```

```python
#measuring the accuracy of training data
log_reg.score(X_train_scaled,y_train)
```

```python
#measuring the accuracy of test data
log_reg.score(X_test_scaled,y_test)
```

```python
log_reg1=LogisticRegression(random_state=0,
                           C=1,
                           fit_intercept=True,
                           ).fit(X_train_scaled,y_train)
```

```python
#measure the accuracy
log_reg1.score(X_train_scaled, y_train)
```

```python
#accuracy for test data
log_reg1.score(X_test_scaled, y_test)
```
